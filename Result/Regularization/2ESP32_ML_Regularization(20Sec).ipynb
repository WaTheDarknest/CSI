```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
url='https://raw.githubusercontent.com/WaTheDarknest/CSI/main/Dataset/Time/2ESP32Dataset(20Sec).csv'
# อ่านข้อมูลไฟล์ csv จาก url
df=pd.read_csv(url)
# เเสดง 10 ตัวอย่างเเรก
df.head(10)
```

```python
# เเสดง 10 ตัวอย่างสุดท้าย
df.tail(10)
```

กำหนดคุณลักษณะในการเรียนรู้ เเละ คุณลักษณะของผลลัพธ์ จากนั้นทำการเเบ่งข้อมูลสำหรับการเรียนรู้เเละทดสอบ

```python
feature_cols=['R1', 'I1', 'R2', 'I2', 'R7', 'I7', 'R8', 'I8', 'R9', 'I9', 'R10', 'I10', 'R11', 'I11', 'R12', 'I12', 'R13', 'I13', 'R14', 'I14', 'R15', 'I15', 'R16', 'I16'
       , 'R17', 'I17', 'R18', 'I18', 'R19', 'I19', 'R20', 'I20', 'R21', 'I21', 'R22', 'I22', 'R23', 'I23', 'R24', 'I24', 'R25', 'I25', 'R26', 'I26', 'R27', 'I27'
       , 'R28', 'I28', 'R29', 'I29', 'R30', 'I30', 'R31', 'I31', 'R32', 'I32', 'R33', 'I33', 'R34', 'I34', 'R35', 'I35', 'R36', 'I36', 'R37', 'I37', 'R38', 'I38'
       , 'R39', 'I39', 'R40', 'I40', 'R41', 'I41', 'R42', 'I42', 'R43', 'I43', 'R44', 'I44', 'R45', 'I45', 'R46', 'I46', 'R47', 'I47', 'R48', 'I48', 'R49', 'I49'
       , 'R50', 'I50', 'R51', 'I51', 'R52', 'I52', 'R53', 'I53', 'R54', 'I54', 'R55', 'I55', 'R56', 'I56', 'R57', 'I57', 'R58', 'I58', 'R59', 'I59']
target_col='TXSpot'
X=df[feature_cols]
y=df[target_col]

test_size=0.01
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=test_size, random_state=7)
```

ทดสอบประสิทธิภาพของแบบจำลองโดยการปรับค่า λ

```python
Lambda_acc = []
Lambda = []
for i in range(1, 101, 1):
  mplo = LogisticRegression(C = i/100, penalty='l2')
  mplo.fit(X_train, y_train)
  Lambda_acc.append(mplo.score(X_train, y_train))
  Lambda.append(1/(i/100))
```

แสดงผลการทดสอบโดยพล็อตกราฟเทียบระหว่างความถูกต้องและ λ

```python
import plotly.express as px
fig = px.line(x=Lambda ,y=Lambda_acc)
fig.update_layout(
    title="Regularized logistic regression",
    xaxis_title="$\lambda$",
    yaxis_title="Accuracy",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="black"
    )
)
fig.show()
```

ทดลองเเสดงค่า intercept คือ θ0 , θ1 และ θ2

```python
#mplo = LogisticRegression(C=1e6)
mplo = LogisticRegression( penalty='l2')
mplo.fit(X_train, y_train)
```

```python
# intercept
print(f'intercept = {mplo.intercept_}') # theta0
# coefficient
print(f'coefficient = {mplo.coef_}') # theta1 theta2 theta3
```

ทดลองดูเหมาะสมกับข้อมูลของแบบจำลองที่สร้างขึ้นเมื่อเทียบกับข้อมูลโดยใช้ค่า R2

```python
R_squred_train =mplo.score(X_train, y_train) # accuracy of the model
print(f'R_squared_error_training = {R_squred_train}')
```

```python
mplo = LogisticRegression(C = 1e6, penalty='l2')
mplo.fit(X_train, y_train)
print(f'R_squared_error_training = {mplo.score(X_train, y_train)}')

regmplo = LogisticRegression(C = 0.01, penalty='l2')
regmplo.fit(X_train, y_train)
print(f'R_squared_error_training = {regmplo.score(X_train, y_train)}')
```

K-fold cross validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean
from numpy import std
```

```python
# prepare the cross-validation procedure
cv = KFold(n_splits=5, random_state=1, shuffle=True)
# # evaluate model
scores = cross_val_score(mplo , X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
# # report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

Regularization

```python
model_reg = LogisticRegression(penalty='l2', C=1.0)
model_reg.fit(X_train, y_train)
```

```python
# intercept
print(f'intercept = {model_reg.intercept_}') # theta0
# coefficient
print(f'coefficient = {model_reg.coef_}') # theta1 theta2 theta3
```

```python
# prepare the cross-validation procedure
cv = KFold(n_splits=5, random_state=1, shuffle=True)
# # evaluate model
scores = cross_val_score(model_reg , X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
# # report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
