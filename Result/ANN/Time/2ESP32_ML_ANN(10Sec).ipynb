```python
# เรียกใช้เครื่องมือ
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/WaTheDarknest/CSI/main/Dataset/Time/2ESP32Dataset(10Sec).csv'
# อ่านข้อมูลไฟล์ csv จาก url
df=pd.read_csv(url)
# เเสดง 10 ตัวอย่างแรก
df.head(10)
```

```python
# ปรับช่วงข้อมูลด้วยวิธี standardization
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

# 'I2', 'R3', 'I3', 'R4', 'I4', 'R5', 'I5', 'R6', 'I6', 'R60', 'I60', 'R61', 'I61', 'R62', 'I62', 'R63', 'I63', 'R64', 'I64'
# ไม่นำมาพิจารณาเนื่องจากค่าตัวแปรดังกล่าวของข้อมูลทุกชุดที่ได้จากกการเก็บตำแหน่งมีค่าเป็น 0 ทั้งชุดข้อมูล
cols=['R1', 'I1', 'R2', 'I2', 'R7', 'I7', 'R8', 'I8', 'R9', 'I9', 'R10', 'I10', 'R11', 'I11', 'R12', 'I12', 'R13', 'I13', 'R14', 'I14', 'R15', 'I15', 'R16', 'I16'
       , 'R17', 'I17', 'R18', 'I18', 'R19', 'I19', 'R20', 'I20', 'R21', 'I21', 'R22', 'I22', 'R23', 'I23', 'R24', 'I24', 'R25', 'I25', 'R26', 'I26', 'R27', 'I27'
       , 'R28', 'I28', 'R29', 'I29', 'R30', 'I30', 'R31', 'I31', 'R32', 'I32', 'R33', 'I33', 'R34', 'I34', 'R35', 'I35', 'R36', 'I36', 'R37', 'I37', 'R38', 'I38'
       , 'R39', 'I39', 'R40', 'I40', 'R41', 'I41', 'R42', 'I42', 'R43', 'I43', 'R44', 'I44', 'R45', 'I45', 'R46', 'I46', 'R47', 'I47', 'R48', 'I48', 'R49', 'I49'
       , 'R50', 'I50', 'R51', 'I51', 'R52', 'I52', 'R53', 'I53', 'R54', 'I54', 'R55', 'I55', 'R56', 'I56', 'R57', 'I57', 'R58', 'I58', 'R59', 'I59']

# กำหนดตัวเเปร X เพื่อเก็บเป็น features ที่ต้องการ
X=df[cols]
# สร้าง object ชื่อ scaler จาก class ของ StandardScaler เพื่อใช้สำหรับการเรียกใช้ในการปรับช่วงข้อมูล
scaler = StandardScaler()
# ปรับช่วงข้อมูลเเละเก็บอยู่ในตัวเเปร X_scale ซึ่งข้อมูลจะอยู่ในรูปของ numpy array
X_scale=scaler.fit_transform(X)
# ปรับให้เเสดงด้วยตัวเลขทศนิยม 0 ตำเเหน่ง
X_scale = X_scale.round(0)
y = np.array(df['TXSpot'])

encoder = OneHotEncoder()
y_targets = encoder.fit_transform(df[['TXSpot']]).toarray()
```

```python
# เเบ่งชุดข้อมูลออกเป็นชุดเรียนรู้เเละชุดทดสอบ
from sklearn.model_selection import train_test_split
test_size=0.01
X_train, X_test, y_train, y_test = train_test_split(
    X, y_targets,
    test_size=test_size, random_state=4)
```

```python
from keras.models import Sequential
from keras.layers import Dense
# สร้างแบบจำลอง
model = Sequential()

# Nh = Ns / (a*(Ni+No)) = (1-0.01)*2217 / (2*(110+3)) = 9.9482 =10

# เนื่องจากใช้ Feature 110 Feature นำมาพิจารณา จึงกำหนดให้ input_dim=110
# สร้าง hidden layer ที่ 1 ประกอบด้วย 10 โหนด ซึ่งเชื่อมต่อกับ Input layer ประกอบด้วย 110 โหนด
# เเละเลือกใช้ ReLu เป็น activation function
model.add(Dense(10, input_dim=110, activation='relu'))

# สร้าง hidden layer ที่ 2 ประกอบด้วย 10 โหนด ซึ่งเชื่อมต่อกับ Input layer ประกอบด้วย 10 โหนด
# เเละเลือกใช้ ReLu เป็น activation function
model.add(Dense(10, input_dim=10, activation='relu'))

# สร้าง hidden layer ที่ 3 ประกอบด้วย 10 โหนด ซึ่งเชื่อมต่อกับ Input layer ประกอบด้วย 10 โหนด
# เเละเลือกใช้ ReLu เป็น activation function
model.add(Dense(10, input_dim=10, activation='relu'))

# Class ของ Feature ในที่นี้ถูกแยกด้วย TXSpot มีทั้งหมด 3 Class จึงกำหนดให้ output_dim=3
# สร้าง output layer ประกอบด้วย 3 โหนด ซึ่งเชื่อมต่อกับ hidden layer ที่ 3 ประกอบด้วย 10 โหนด
# เเละเลือกใช้ sigmoid เป็น activation function
model.add(Dense(3, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, verbose=1)
```

```python
# ทดสอบประสิทธิภาพแบบจำลอง
results = model.evaluate(X_test, y_test)
print('Accuracy: %.3f' % results[1])
```
